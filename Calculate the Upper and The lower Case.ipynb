```python
def string_test(s):
    d={"x":0, "y":0}
    for c in s:
        if c.isupper():
           d["x"]+=1
        elif c.islower():
           d["y"]+=1
        else:
           pass
   
    print ("Number of Upper case characters : ", d["x"])
    print ("Number of Lower case Characters : ", d["y"])

string_test('The Edyoda University')
```
